```python
from sqlalchemy import create_engine
from sqlalchemy.orm import sessionmaker
import pandas as pd
import numpy as np
import pymysql
```

```python
webtoon_db_connect_str = "mysql+pymysql://jsh:bread!123@localhost/jsh"
webtoon_db_connection = create_engine(webtoon_db_connect_str)
Session = sessionmaker(webtoon_db_connection)
conn = webtoon_db_connection.connect()
```

```python
survey_df = pd.read_csv("c:/kth/webtoon/survey_db_data.csv",index_col=0)
webtoon_info_df = pd.read_csv("c:/kth/webtoon/drop_webtoon_preprocessing_221129.csv",index_col=0,encoding="cp949")

# survey_df에 "webtoon_no" 컬럼 추가
for survey_index in range(len(survey_df)):
    for webtoon_info_index in range(1,len(webtoon_info_df)):
        if webtoon_info_df.loc[webtoon_info_index,"title"] == survey_df.loc[survey_index,"webtoon_title"]:
            survey_df.loc[survey_index,"webtoon_no"] = webtoon_info_df.index[(webtoon_info_df["title"]==survey_df.loc[survey_index,"webtoon_title"])]

survey_df = survey_df.astype({"webtoon_no":"int"})
survey_df = survey_df.drop("webtoon_title",axis=1)
survey_df = survey_df[["user","webtoon_no","score"]]

survey_df.to_sql(name="survey",con=webtoon_db_connection,if_exists="append",index=False)

with Session() as session:
    session.commit()
```
