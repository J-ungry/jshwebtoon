# 데이터 전처리 
## ages
    + 전체이용가 -> 0
    + 12세 -> 1
    + 15세 -> 2
    + 18세 -> 3
## rates
    + float 형 변환  소수점 두개만 남기고 잘라버리기 
## genre
    - genre1
        + 스토리 -> 0
        + 에피소드 -> 1
        + 옴니버스 -> 2
    - genre2
        + 로맨스 ->0
        + 드라마 ->1
        + 판타지 ->2 
        + 스릴러 ->3 
        + 액션  ->4
        + 개그  ->5 
        + 일상  ->6 
        + 무협/사극   ->7  
        + 감성  ->8 
        + 스포츠  ->9 
       
## likes 
    99,999 + 를 100000 으로 변환
    전체 Int 형으로 변환

```python
import pandas as pd
import numpy as np

df = pd.read_csv("./webtoon_221128.csv",index_col=0)
```

```python
df
```

```python
df['ages'].value_counts()
```

```python
#ages 전처리
df['ages2'] = df['ages'].map({"전체연령가":0,"12세 이용가":1,"15세 이용가":2,"18세 이용가":3})
```

```python
df['ages2'].value_counts()
```

```python
df.head(10)
```

```python
#rates 전처리 (float로 바꾸고 소수점 아래 2개까지만)
rates2 = []
for x in df['rates']:
   rates2.append(format(x,".2f"))
df['rates2'] = rates2
df = df.astype({'rates2':'float64'})
```

```python
#genre 전처리
genre1 = []
genre2 = []
for gs in df['genres']:
    genre1.append(gs.replace(' ', '').split(',')[0])
    genre2.append(gs.replace(' ', '').split(',')[1])
df['genre1'] = genre1
df['genre2'] = genre2
```

```python
print(df['genre1'].value_counts())
print(df['genre2'].value_counts())
```

```python
df['genre1_pre'] = df['genre1'].map({'스토리':0,'에피소드':1,"옴니버스":2})
df['genre2_pre'] = df['genre2'].map({'로맨스':0,'드라마':1,"판타지":2,"스릴러":3,"액션":4,"개그":5,"일상":6,"무협/사극":7,"감성":8,"스포츠":9})
```

```python
print(df['genre1_pre'].value_counts())
print(df['genre2_pre'].value_counts())
```

```python
#likes 전처리 
df['likes'] = df['likes'].replace('99,999+','100000')
for x in range(0, len(df['likes'])):
    df.loc[x,'likes'] = df.loc[x,'likes'].replace(',', '')
df['likes'] = df['likes'].replace('좋아요','0')
df['likes'] = pd.to_numeric(df['likes'])
```
