```python
import pandas as pd
import numpy as np
```

```python
rating_data = pd.read_csv("../../csv/after_pre/total_survey_221129.csv",encoding="cp949",index_col=0)
webtoon_data = pd.read_csv("../../csv/after_pre/drop_webtoon_preprocessing_221130.csv",encoding="cp949",index_col=0)
```

```python
rating_data = rating_data.fillna(0)
rating_data = rating_data.drop(["gender","age"],axis=1)
rating_data
```

```python
webtoon_title=webtoon_data["title"]
webtoon_title_list=list(webtoon_title)
```

```python
webtoon_user_pivot = rating_data.values.T
webtoon_user_pivot
```

```python
from sklearn.decomposition import TruncatedSVD

SVD = TruncatedSVD(n_components=12)
matrix = SVD.fit_transform(webtoon_user_pivot)
matrix.shape
```

```python
matrix[8]
```

```python
corr = np.corrcoef(matrix)
corr.shape
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(16,10))
sns.heatmap(corr)
```

```python
def recommand_webtoon(title):
    target=webtoon_title_list.index(title)
    corr_target=corr[target]
    result=list(webtoon_title[(corr_target>=0.8)])[:50]
    return result
```

```python
recommand_webtoon("마음의소리")
```
