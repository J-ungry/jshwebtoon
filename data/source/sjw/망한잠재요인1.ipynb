```python
import pandas as pd
import numpy as np
```

```python
rating_data = pd.read_csv("../../csv/after_pre/total_survey_221129_2.csv")
webtoon_data = pd.read_csv("../../csv/after_pre/drop_webtoon_preprocessing_221130.csv",encoding="cp949",index_col=0)
```

```python
rating_data = rating_data.rename(columns={"webtoon_title":"title"})
```

```python
rating_data
```

```python
#webtoon_title_list = list(webtoon_data['title'])
```

```python
user_webtoon_data = pd.merge(webtoon_data,rating_data,on="title")
```

```python
user_webtoon_data
```

```python
user_webtoon_data = user_webtoon_data.drop('status',axis=1)
user_webtoon_data = user_webtoon_data.drop('intro',axis=1)
user_webtoon_data = user_webtoon_data.drop('episodes',axis=1)
user_webtoon_data = user_webtoon_data.drop('first_register_date',axis=1)
user_webtoon_data = user_webtoon_data.drop('last_register_date',axis=1)
user_webtoon_data = user_webtoon_data.drop('age',axis=1)
user_webtoon_data = user_webtoon_data.drop('rate',axis=1)
user_webtoon_data = user_webtoon_data.drop('likes',axis=1)
```

```python
user_webtoon_data
```

```python
user_webtoon_pivot = user_webtoon_data.pivot_table("webtoon_ratings",index='user',columns='title')
```

```python
user_webtoon_pivot = user_webtoon_pivot.fillna(0)
```

```python
user_webtoon_pivot
```

```python
user_webtoon_pivot.shape
```

```python
webtoon_user_pivot = user_webtoon_pivot.values.T
webtoon_user_pivot.shape
```

```python
from sklearn.decomposition import TruncatedSVD

SVD = TruncatedSVD(n_components=12)
matrix = SVD.fit_transform(webtoon_user_pivot)
matrix.shape
```

```python
matrix[0]
```

```python
import numpy as np
corr = np.corrcoef(matrix)
corr.shape
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(16,10))
sns.heatmap(corr)
```

```python
webtoon_title = webtoon_data.columns
print(webtoon_title)
webtoon_title_list= list(webtoon_title)
print(webtoon_title_list)
```

```python
webtoon_title
```

```python
def recommend_webtoon(title):
    target = webtoon_title_list.index(title)
    corr_target = corr[target]
    result = list(webtoon_title[(corr_target>=0.9)])[:200]
    
    return result
```
